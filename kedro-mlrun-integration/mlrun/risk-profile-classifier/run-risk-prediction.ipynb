# Risk Profile Prediction - Nuclio Function

A simple Nuclio function that generates random risk profile predictions based on user name.


```python
import mlrun

project_name = "kedro-test"
project = mlrun.get_or_create_project(project_name)
```

```python
# Create nuclio function from the user_info.py file
function = project.set_function(
    kind="nuclio",
    name="get-user-info",
    func="user_info.py",
    handler="handler",
    image="python:3.11"
)
```

```python
function.deploy()
```
