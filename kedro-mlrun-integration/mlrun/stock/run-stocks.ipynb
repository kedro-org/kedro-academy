```python
import mlrun
project_name = "stocks"
```

```python
project = mlrun.get_or_create_project(project_name)
```

```python
function = project.set_function(kind="remote", name="stock-price", func="stocks.py")
function.with_requirements([
    'yfinance',
    'pandas',
    'cacheout'
])
```

```python
function.set_env("CACHE_TTL_SECONDS", "300")
```

```python
function.deploy()
```
