```python
import mlrun
project_name = "tax-expert"
```

```python
project = mlrun.get_or_create_project(project_name)
```

```python
function = project.set_function(kind="remote", name="bank-agent", requirements=["langchain_nvidia_ai_endpoints", "langchain-openai"], func="longchain-tax-expert.py", image="mlrun/mlrun")
function.spec.nuclio_runtime="python:3.9"
function.with_requirements([
    'langchain-core>=0.1.0',
    'langchain>=0.1.0',
    'langchain-community>=0.0.20',
    'langchain-nvidia-ai-endpoints>=0.1.0',
    'faiss-cpu>=1.7.4',
    'pydantic>=2.0.0'
])
```

```python
MODEL_NAME = "meta/llama3-8b-instruct"
```

```python
function.set_env("MODEL_NAME", MODEL_NAME)
function.set_env("NIM_URL", "nuclio-mlrun-nvidia-nim-demo-kate-my-nim-kate.default-tenant.svc.cluster.local:8080")
```

```python
function.deploy()
```
